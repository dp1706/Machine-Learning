# ***PART 1: Single layer perceptron(Slp)***

```
#learning rates

alpha_1 =0.0001
alpha_2 = 0.001
alpha_3 = 0.01
alpha_4 = 0.1
alpha_5 =0.5
```

```
#creating weight in range -0.3 to 0.3 
def weight(): 
    weigth_1=[]
    for i in range(3): 
        weigth_1.append(round(random.uniform(-0.3,0.3),2))
    return weigth_1
```

```
#threshold Activation Function

def Activation(input,weight):
  total_sum=np.dot(input,weight[1:])+weight[0]
  if total_sum>0:
    activation=1
  else :
    activation=0
  return activation
```

```
#Train model

def Train_model(input,output,weight,epoch,learning_rate):
  error_s = []
  for _ in range(epoch):
    error = 0
    for inputs,outputs in zip(input,output):
      activation=Activation(inputs,weight)
      sum = learning_rate * (outputs-activation)
      error +=(1- (outputs-activation) * (outputs-activation))/4
      weight[1:]+=sum * inputs 
      weight[0]+=sum
    error_s.append(error)
  return weight,error_s
```

```
#for and Gate

input=np.array([[1,1],[1,0],[0,1],[0,0]])
output=np.array([1,0,0,0])

epoch=1000
#Training the model
weight1 = weight()
print("weight is : ", weight1)
new_weight , error =Train_model(input,output,weight1,epoch,alpha_1)
print("new weight is : ",new_weight)

#testing phase
new_input=np.array([0,1])
ans=Activation(new_input,new_weight)
print(ans)

epochs = range(0,1000)
plt.title("training vs epochs")
plt.plot(epochs,error,color='r',label='accuracy line')
plt.xlabel("epochs")
plt.ylabel("training accuracy")
plt.legend()
plt.show()
```

```
"""
train set and test set """

def splitDataset(dataset,splitratio):
  size = int(len(dataset) * splitratio)
  x_train = []
  copy=list(dataset)
  while len(x_train) < size :
    index = random.randrange(len(copy))
    x_train.append(copy.pop(index))
  return [x_train,copy]    
```

```
""" spliting dataset
     with the ratio of splitratio 
"""
splitratio=0.2
trainSet ,testSet = splitDataset(mydata,splitratio)
```

```
"""
seprating dataset by class

here we will get

2 class :

malignant, benign
"""

def sepratebyclass(dataset):
  separated = {}
  for i in range(len(dataset)):
    vector = dataset[i]
    if ( vector[-1] not in separated):
        separated[vector[-1]] = []
    separated[vector[-1]].append(vector)
  return separated
```

```
"""
for continuous data we are using gaussian distribution 
so we need mean and std
"""

def Mean(number):
  return float(sum(number))/float(len(number))
```

```
#standard daviation

def Stddev(number):
  avg=Mean(number)
  variance=sum([(x-avg)**2 for x in number])/float(len(number)-1)
  return sqrt(variance)
```

```
"""
storing mean and stddev for each feature column
"""

def summarize(dataset):
  srs = [(Mean(column), Stddev(column), len(column)) for column in zip(*dataset)]
  del srs[-1]
  return srs
```

```
"""
storing data of each seprated class
"""

def summarize_by_class(dataset):
  seperated = sepratebyclass(dataset)
  sumrs = {}
  for classvalue , instances in seperated.items():
     sumrs[classvalue]=summarize(instances)
  return sumrs
```

```
"""
calculating gaussian distribution pdf

"""

def calculate_probability(x,mean,stddev):
  exponent = math.exp(-((x-mean)**2/(2 * stddev**2)))
  return (1/(sqrt(2*math.pi)* stddev)) * exponent
```

```
# calculating class probability


def calculate_class_probability(summaries , input):
  probabilities = {}
  for classvalue , class_summaries in summaries.items():
   probabilities[classvalue]=1
   for i in range(len(class_summaries)):
     mean , stddv , lenght =class_summaries[i]
     x=input[i]
     probabilities[classvalue] *= calculate_probability(x,mean,stddv)
   return probabilities  
```

```
"""
predicting probability for test datasets
"""

def predict(summaries,input):
  probabilities = calculate_class_probability(summaries,input)
  bl,bp=None,-1
  for classvalue,probability in probabilities.items():
    if bl is None or probability > bp:
      bp=probability
      bl=classvalue
  return bl
```

```
#caller function for predict

def bays_classifier(summaries,testSet):
  prediction = []
  for i in range(len(testSet)):
    result = predict(summaries,testSet[i])
    prediction.append(result)
  return prediction
```

```
"""
calculating accuracy
"""

def Accuracy(testSet,prediction):
  correct =0
  for x in range(len(testSet)):
    if testSet[x][-1] == prediction[x]:
      correct += 1
  return (correct/float(len(testSet)))*100.0
```

```
"""
here we traning out model and testing and finding Accuracy

trainset consist of  x_train and Y_train  as trainset[x][-1] is the value of Y_train.

similar testSet consist of X_test,Y_test

"""

summaries = summarize_by_class(trainSet)

predictions = bays_classifier(summaries,testSet)

accuracy = Accuracy(testSet,predictions)

print("Accuracy : {0}%".format(accuracy))
```
